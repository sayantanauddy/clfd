# HelloWorld Dataset

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_dir = 'datasets/robot_hello_world/processed_demos'
sequence_file = 'datasets/robot_hello_world/helloworld_sequence_all.txt'

fig, ax = plt.subplots(1,7,figsize=(30,4), sharey=True, sharex=True)
with open(sequence_file, 'r') as seq_file:
    for i,filename in enumerate(seq_file.readlines()):
        filename = filename.replace('\n','')
        data_file = os.path.join(dataset_dir, filename)
        data = np.load(data_file)
        data_shape = data.shape
        for demo in range(data_shape[0]):
            ax[i].plot(data[demo, :, 0], data[demo, :, 1], lw=4, alpha=0.6)
        title = f'{filename} {data_shape}'
        ax[i].set_title(title, fontsize=20)
        ax[i].get_xaxis().set_visible(False)
        ax[i].get_yaxis().set_visible(False)
        ax[i].xaxis.get_label().set_visible(False)
        ax[i].yaxis.get_label().set_visible(False)
        ax[i].invert_xaxis()
        ax[i].invert_yaxis()

fig.subplots_adjust(wspace=0.1)
plt.savefig('videos_images/HelloWorld_dataset.svg', bbox_inches='tight')
```
